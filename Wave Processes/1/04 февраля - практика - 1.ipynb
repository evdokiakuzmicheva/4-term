# Основы численных методов

```python
# Импорт используемых модулей
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Рисуем аналитическое решение u(t) = exp(-t), вводя разностную сетку (т.к. на компьютере). Уравнение du/dt=-u(t).
t_n = np.linspace(0., 5., 11)
u_n = np.exp(-t_n)
plt.plot(t_n, u_n, label="Аналитическое решение u(t)")
plt.legend()
plt.show()
```

```python
# Рассчитываем разностное решение U(t_n+1) = U(t_n) + dt * U(t_n) на той же сетке, явная схема Эйлера
T = 5
N = 11
t_n = np.linspace(0., T, N)
dt = t_n[1] - t_n[0]
U_n = np.zeros(N)
U_n[0] = 1.
for idx in np.arange(1, N):
    U_n[idx] = U_n[idx-1] - dt * U_n[idx-1]
plt.plot(t_n, U_n, 'o', label="Разностное решение U(t)")
plt.legend()
plt.show()
```

```python
# Накладываем два графика, чтобы проиллюстрировать O(h) ошибку
plt.plot(t_n, u_n, label="Аналитическое решение u(t)")
plt.plot(t_n, U_n, 'o', label="Разностное решение U(t)")
plt.legend()
plt.show()
```
